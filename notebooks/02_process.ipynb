```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib
df=pd.read_csv('../data/raw.csv')
```

```python
le=LabelEncoder()
df['Result_Encoded']=le.fit_transform(df['Result'])
```

```python
df.drop(columns='Result',inplace=True,axis=1)
```

```python
df.to_csv('../data/processed.csv',index=False)
```
