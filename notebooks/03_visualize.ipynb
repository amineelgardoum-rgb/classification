```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../data/processed.csv')
df
```

```python
df['Systolic blood pressure'].plot(kind='hist',bins=10,edgecolor='black')
plt.xlabel('Systolic blood pressure')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Age'].plot(kind='hist', bins=10, edgecolor='black')
plt.xlabel('Age')
plt.show()
```

```python
df['Troponin'].plot(kind='hist',bins=10)
plt.xlabel("Troponin")
plt.ylabel("Frequency")
plt.title("Troponin Vs frequency")
plt.show()
```

```python
plt.scatter(df['Age'],df['Heart rate'],color='Red')
plt.title("Age vs Heart rate")
plt.xlabel('age')
plt.ylabel('heart rate')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(df.corr(),annot=True,fmt='.2f',cmap='inferno',linewidths=1)
plt.title("Heatmap of Correlation")
plt.show()
```
